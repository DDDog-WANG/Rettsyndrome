# Uint16 to Uint8

```python
def NormToUnit8(img):
    img_uint8 = img - img.min()
    img_uint8 = img_uint8 / (img_uint8.max() - img_uint8.min())
    img_uint8 *= 255
    return img_uint8.astype(np.uint8)
```

```python
def NormToUnit8(img):
    img_uint8 = img
    img_uint8[img_uint8>255]=255
    return img_uint8.astype(np.uint8)
```
